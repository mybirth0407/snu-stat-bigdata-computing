### What is UNIX?

#### History of Unix
![history-of-unix](https://upload.wikimedia.org/wikipedia/commons/thumb/7/77/Unix_history-simple.svg/1920px-Unix_history-simple.svg.png)

UNIX는 교육 및 연구 기관에서 가장 활발히 사용되어지고 있는 운영체제의 한 종류입니다. 1970년대의 벨 연구소에서 켄 톰슨, 데니스 리치 등이 **소형 컴퓨터를 위한 이식성 높은 운영체제**를 개발하는 것으로부터 시작되었습니다. 최근에는 UNIX 시스템을 기반으로 하여 여러 회사들과 영리/비영리 단체들이 그들에게 필요한 운영체제를 만들고 사용자에게 제공하고 있습니다. UNIX는 설계 단계부터 다양한 시스템 사이의 이식성을 고려하여 개발되었고, 그 결과 [Android OS](https://en.wikipedia.org/wiki/Android_(operating_system)), [iPhone OS](https://en.wikipedia.org/wiki/IOS) 와 같이 우리에게 친숙한 형태로 제공되고 있습니다.

사실 엄밀하게는, 우리에게 제공되고 있는 많은 UNIX 시스템은 UNIX가 아닌 UNIX-like 시스템입니다. UNIX-like한 운영체제로는 앤드류 타넨바움이 교육 목적으로 제작한 Minix와 리누스 토르발즈가 취미로 제작한 Linux가 있습니다. 두 운영체제 모두 오픈소스 프로젝트입니다. 그 중, 현재 클라우드 컴퓨팅 점유율 90%, 스마트폰의 82%, 임베디드 기기의 62%, 슈퍼 컴퓨터 시장의 99%를 차지하고 있는 것은 Linux 입니다. 앞서 언급한 Android OS가 Linux 기반 운영체제에 해당합니다.<br>
반면, 정통 UNIX 계열의 운영체제도 최근 10년간 상승세를 타고 있습니다. 바로 Apple 덕분입니다. Apple의 운영체제들은 정통 UNIX 계열인 BSD의 $^*$*커널* 을 활용하고 있습니다. 여담으로, Apple이 사용하는 커널의 명칭은 XNU (**X** is **N**ot **U**NIX)인데, UNIX 인증을 받을 정도로 정통 UNIX 시스템입니다.

*: 운영체제의 핵심이 되는 프로그램으로, 운영체제의 모든 것에 접근할 수 있으며 완전한 제어가 가능하다. 하드웨어와 프로세스의 보안부터, 응용 프로그램으로의 적절한 자원 분배 등의 역할을 한다.

### Shell

#### What is a Shell?
*쉘 (Shell)* 쉘은 사용자 입력을 읽어서 명령을 실행하는 $^*$*명령 줄 해석기 (CLI)* 입니다. 쉘은 껍데기를 뜻하는데, 사용자 입장에서 보면 커널을 감싸고 있기 때문에 쉘이라는 이름이 붙었습니다. 쉘은 일반적으로 $^{**}$*문자 기반 사용자 인터페이스 (CLI)* 로 분류되며, Windows에서 주로 활용되는 *$^{***}$그래픽 사용자 인터페이스 (GUI)* 와는 대조되는 사용자 환경입니다. 쉘에 대한 사용자 입력은 보통 터미널 (대화형 쉘)을 통해 주어지지만, 명령어 모음을 파일로 만든 *쉘 스크립트 (shell script)* 로 주어질 수도 있습니다. 사용자가 쉘을 통해 명령을 입력하면, 쉘은 커널을 통해 하드웨어 연산을 요청하고, 연산 결과를 커널을 통해 전달받아 사용자에게 출력해줍니다.

### File and Directory

#### File system
UNIX 파일 시스템은 디렉토리들과 파일들을 *계층적 (hierarchical)* 으로 조직화한 것입니다. 파일 시스템의 모든 것은 *루트 디렉토리 (root)* 부터 시작한다. 루트 디렉토리는 **슬래시 (/)** 로 표시합니다. 디렉토리는 *디렉토리 항목 (directory entry)* 를 담고 있는 *파일 (file)* 입니다. $^*$논리적으로, 하나의 디렉토리 항목은 *파일명* 과 *파일의 특성* 을 기술하는 정보 구조체를 담고 있는 무언가라고 생각하면 됩니다. 파일의 특성으로는 *파일의 종류*, *파일의 크기*, *파일의 소유자 및 권한* 등이 있으며, **stat** 과 같은 *명령어 (command)* 로 파일의 특성을 확인할 수 있습니다.<br>
실습 시간에 직접 명령어를 실행하여 파일의 특성들을 확인해볼 것입니다.


*: 실제 디스크에 저장되는 자료가 아닌, 소프트웨어 위의 논리 구조를 뜻합니다.

#### File types and permissions
UNIX 파일 시스템에서 파일의 종류와 권한은 10자리의 문자로 표시되며 (예: drwxr-xr-x), 각각은 다음과 같은 의미를 갖습니다.

|Filetype (1)|User (Permission) (3)|Group (3)|Others (3)|
|:---|:---:|:---:|:---:|
|d|rwx|r-x|r-x||

파일의 종류는 한 자리의 문자로 표시되며 8종류로 구별됩니다.
|Character (1)|Type|
|:---|:---:|
|-|Regular file|
|d|Directory|
|D|Door file|
|c|Character device file|
|b|Block device file|
|s|Socket file|
|p|FIFO (Named pipe) file|
|l|Symbolic link file|

파일의 권한은 9자리 문자로 표시되며, 3자리씩 각각 사용자, 그룹, 외부에 해당하는 권한을 표시합니다. 각각의 권한은 동일한 표기법을 이용하며, 자릿수 순서대로 읽기 (r), 쓰기 (w), 실행 (x) 권한을 나타냅니다. 권한을 나타내는 표기법으로는 숫자 표기법과 문자 표기법이 있습니다. 다음 테이블은 첫번째 열에 숫자 표기법을, 나머지 열에 문자 표기법을 표시합니다.
|Numeric notation|Read (4)|Write (2)|Execute (1)|Description|
|:---:|:---:|:---:|:---:|:---:|
|0|-|-|-|No permissions|
|1|-|-|x|**Only execute**|
|2|-|w|-|**Only write**|
|3|-|w|x|Write and execute|
|4|r|-|-|**Only read**|
|5|r|-|x|Read and execute|
|6|r|w|-|Read and write|
|7|r|w|x|Read, write, and execute|

#### Filename
하나의 디렉토리 안의 디렉토리 항목들의 이름을 *파일명 (filename)* 이라고 부릅니다. 파일명에는 대부분의 문자를 사용할 수 있습니다. 사용할 수 없는 문자는 단 두 개로, *슬래시 (/)* 와 *NULL* 문자입니다. 슬래시는 하나의 *경로명 (pathname)* 을 구성하는 파일명들을 구분하는 역할을 하고, NULL 문자는 경로명의 끝을 표시합니다. NULL 문자는 컴퓨터 공학 관점에서 중요한 역할을 하지만, 본 부트캠프에서는 일단 *사용하기에 불편한 문자* 정도로만 기억합시다. 그럼에도 불구하고, 파일명은 보통 $^*$사용자가 이해하기 편한 문자들만 사용하는 것이 권장됩니다. 파일명에 특수문자들이 포함되어 있으면 사용자의 언어 환경에 따라 읽고 쓰는 것이 상당히 복잡해질 수 있습니다.


*: 운영체제 (혹은 시스템)간의 이식성을 위해 $^{**}$[POSIX](http://get.posixcertified.ieee.org/)는 파일명에 사용할 수 있는 문자를 영어(a-z, A-Z), 숫자(0-9), 마침표(.), 대쉬(-), 밑줄(_)로만 제한합니다.<br>
**: POSIX: [IEEE (전기전자공학자협회)](https://www.ieee.org/)가 제정한 UNIX [응용 프로그래밍 인터페이스 (API)](https://en.wikipedia.org/wiki/API) 규격입니다.

#### Pathname
경로명(pathname)은 하나 이상의 파일명들이 슬래시로 연결된 형태입니다. 경로명의 첫 글자가 / 일 수도 있으며, 이 경우 루트 디렉토리 이하의 모든 경로를 표시하는 것입니다. 이러한 표시 방식을 *절대 경로 (absolute path)* 혹은 *전체 경로 (full path)* 라고 합니다. 그렇지 않은 경우를 *상대 경로 (relative path)* 라고 하며, 이것은 현재 사용자의 디렉토리로부터 시작하는 경로의 상대적인 파일을 지칭합니다.

#### Special path
UNIX에는 새 디렉토리를 만들면 자동으로 생성되는 파일명이 두 개 있습니다. 하나는 **.**, 다른 하나는 **..** 입니다. .는 현재 디렉토리를 뜻하고, ..는 부모 디렉토리를 뜻합니다. 루트 디렉토리에서는 .와 ..가 동일하게 자기 자신인 루트 디렉토리를 뜻합니다. **~** 경로도 자주 사용됩니다. ~는 사용자의 홈 디렉토리입니다.

삭제하고자 하는 디렉토리가 비어있는 경우 단일 명령어로 삭제할 수 있습니다.

````python
!ls
````

대부분의 경우, 삭제하고자 하는 디렉토리가 비어있지 않은 경우가 많으므로 rm -r 명령어를 사용하는 것이 더 편리합니다. 반면, 삭제하고자 하는 파일이 핵심 파일이라면 (시스템 디렉토리, 읽기 전용 파일 등) rm 명령어는 1) 경고 메세지를 출력하거나, 2) 삭제 여부를 확인하거나, 3) 삭제를 거부할 수도 있습니다. 이 경우 강제로 삭제하는 옵션 -f 를 주어 삭제합니다. (--force)
