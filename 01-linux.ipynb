## 01. Linux (and UNIX)


### What is UNIX?

#### History of Unix
![history-of-unix](https://upload.wikimedia.org/wikipedia/commons/thumb/7/77/Unix_history-simple.svg/1920px-Unix_history-simple.svg.png)

*UNIX* 는 교육 및 연구 기관에서 가장 활발히 사용되어지고 있는 운영체제의 한 종류입니다. 1970년대의 벨 연구소에서 켄 톰슨, 데니스 리치 등이 **소형 컴퓨터를 위한 이식성 높은 운영체제**를 개발하는 것으로부터 시작되었습니다. 최근에는 UNIX 시스템을 기반으로 하여 여러 회사들과 영리/비영리 단체들이 그들에게 필요한 운영체제를 만들고 사용자에게 제공하고 있습니다. UNIX는 설계 단계부터 다양한 시스템 사이의 이식성을 고려하여 개발되었고, 그 결과 [Android OS](https://en.wikipedia.org/wiki/Android_(operating_system)), [iPhone OS](https://en.wikipedia.org/wiki/IOS) 와 같이 우리에게 친숙한 형태로 제공되고 있습니다.

사실 엄밀하게는, 우리에게 제공되고 있는 많은 시스템은 UNIX가 아닌 $^{*}$ UNIX-like 시스템입니다. UNIX-like한 운영체제로는 앤드류 타넨바움이 교육 목적으로 제작한 *Minix* 와 리누스 토발즈가 취미로 제작한 *Linux*가 있습니다. 두 운영체제 모두 오픈소스 프로젝트입니다. 그 중, 현재 클라우드 컴퓨팅 점유율 90%, 스마트폰의 82%, 임베디드 기기의 62%, 슈퍼 컴퓨터 시장의 99%를 차지하고 있는 것은 Linux 입니다. 앞서 언급한 *Android OS* 가 Linux 기반 운영체제에 해당합니다.<br>
반면, 정통 UNIX 계열의 운영체제도 최근 10년간 상승세를 타고 있습니다. 바로 Apple 덕분입니다. Apple의 운영체제들은 정통 UNIX 계열인 BSD의 $^{**}$*커널* 을 활용하고 있습니다. 여담으로, Apple이 사용하는 커널은 Mach 라는 커널 위에 BSD 의 일부 요소를 더하여 만든 *XNU* (**X** is **N**ot **U**NIX)인데, 이름과 달리 UNIX 인증을 받을 정도로 정통 UNIX 시스템입니다.

$^{*}$: [TMI: UNIX 핵심 개발자 데니스 리치는 Unix-like 운영체제를 사실상 UNIX 라고 생각한다는 인터뷰를 했습니다.](http://www.linuxfocus.org/English/July1999/article79.html)<br>
$^{**}$: 운영체제의 핵심이 되는 프로그램으로, 운영체제의 모든 것에 접근할 수 있으며 완전한 제어가 가능합니다. 하드웨어와 프로세스의 보안부터, 응용 프로그램으로의 적절한 자원 분배 등의 역할을 합니다.

#### Why Linux?
[Differences-between-linux-and-unix](https://opensource.com/article/18/5/differences-between-linux-and-unix)


벨 연구소에서 개발한 UNIX 시스템은 운영체제와 시스템 소프트웨어에 큰 영향을 미쳤습니다. 다만, 당시의 UNIX는 비싼 하드웨어 스펙을 요구하고, 대형 연구소들이 주로 사용했기에 일반 사용자가 사용하기에는 접근성이 좋지 않았습니다. 이러한 UNIX의 단점을 해소하고, 평범한 사용자들에게 UNIX-like 한 시스템을 제공하고자 만들어진 것이 *Linux* 입니다.

리누스 토발즈는 대학생 시절, UNIX 시스템의 종류 중 하나인 *Minix* 를 기반으로 Linux 를 개발했는데, Linux는 모든 소스를 공개하는 오픈소스 소프트웨어 방식으로 개발되었습니다. 이 방식에 공감하는 많은 개발자들이 온라인으로 모여 Unix-like한 환경을 제공하는 Linux 개발에 힘을 모았습니다. 사실 이 전에도 UNIX의 복제본을 만드려는 시도가 있었습니다. 상업적인 UNIX 시스템을 반대하던 리처드 스톨만은 *자유 소프트웨어 재단* 을 설립하고, *GNU* (**G**NU's is **N**ot **U**NIX) 라는, UNIX 시스템의 복제를 만들고 있었는데, 많은 부분을 성공적으로 개발했지만 핵심 커널을 개발하는 데에 실패했습니다. 이 과정에서 리누스 토발즈의 UNIX-like 한 커널을 GNU 시스템에 결합해 만들어진 운영체제가 *GNU/Linux* 현재 우리가 Linux라고 부르는 시스템입니다. 오픈 소스 커뮤니티와 함께 Linux 시스템의 성공에 힘입어 많은 과학자, 공학자, 연구자들이 Linux를 이용하여 연구개발을 하기 시작했고, 현재는 수많은 프로그램들이 Linux에서 동작하며 가장 널리 쓰이는 운영체제로 성장했습니다.

Linux는 오픈소스 커뮤니티 기반으로 발전되었기에, 문제점이 발생했을 때에 수많은 사람들의 참여로 빠르게 해소되어왔습니다. 그러한 과정 덕분에 다양하고 안정적인 환경을 제공합니다. 최근에는 대다수의 수치계산, 혹은 기계학습이나 블록체인 같은 거대한 시스템들이 Linux 기반으로 구축되고 있습니다.

1. **Linux는 오픈 소스 운영체제** 입니다. Windows와 Mac OS X와는 다르게 상업성이 없는 Linux 배포판은 무료로 사용할 수 있습니다.

2. **Linux는 커스터마이징이 자유** 롭습니다. Linux는 앞서 언급했듯, 커널만을 의미합니다. 즉, 운영체제의 핵심 부분만을 제공하고 있으므로, 커널만을 사용하여 운영 체제의 다른 부분들을 사용자의 의도대로 수정하여 사용할 수 있습니다. 대표적으로 [Ubuntu](https://ubuntu.com/), [Linux Mint](https://linuxmint.com/), [CentOS](https://www.centos.org/), 그리고 토발즈가 사용하는 [Fedora](https://fedoraproject.org/) 가 있습니다.

3. **Linux는 안정적** 입니다. 리눅스는 오픈소스로 관리되고 있어, 각계각층의 사용자들이 실시간으로 활용하고 있습니다. 따라서 문제점이 발생할 경우 전 세계 누구나 참여하여 이를 해결하는 데에 기여할 수 있습니다. 즉, 사내 개발만으로 진행되고 있는 운영체제인 Windows와 Mac OS X와는 다르게 빠르게 버그가 수정될 수 있습니다.

### Shell

#### What is a Shell?
*쉘 (Shell)* 쉘은 사용자 입력을 읽어서 명령을 실행하는 *명령 줄 해석기 (CLI, Command Line Interpreter)* 입니다. 쉘은 껍데기를 뜻하는데, 사용자 입장에서 보면 커널을 감싸고 있기 때문에 쉘이라는 이름이 붙었습니다. 쉘은 일반적으로 *문자 기반 사용자 인터페이스 (CUI, Character User Interface)* 로 분류되며, Windows에서 주로 활용되는 *그래픽 사용자 인터페이스 (GUI, Graphical User Interface)* 와는 대조되는 사용자 환경입니다. 쉘에 대한 사용자 입력은 보통 터미널 (대화형 쉘)을 통해 주어지지만, 명령어 모음을 파일로 만든 *쉘 스크립트 (shell script)* 로 주어질 수도 있습니다. 사용자가 쉘을 통해 명령을 입력하면, 쉘은 커널을 통해 하드웨어 연산을 요청하고, 연산 결과를 커널을 통해 전달받아 사용자에게 출력해줍니다.

### File and Directory

#### File system
UNIX 파일 시스템은 디렉토리들과 파일들을 *계층적 (hierarchical)* 으로 조직화한 것입니다. 파일 시스템의 모든 것은 *루트 디렉토리 (root)* 부터 시작합니다. 루트 디렉토리는 **슬래시 (/)** 로 표시합니다. 디렉토리는 *디렉토리 항목 (directory entry)* 를 담고 있는 *파일 (file)* 입니다. $^*$논리적으로, 하나의 디렉토리 항목은 *파일명* 과 *파일의 특성* 을 기술하는 정보 구조체를 담고 있는 무언가라고 생각하면 됩니다. 파일의 특성으로는 *파일의 종류*, *파일의 크기*, *파일의 소유자 및 권한* 등이 있으며, `stat` 과 같은 *명령어 (command)* 로 파일의 특성을 확인할 수 있습니다.<br>
실습 시간에 직접 명령어를 실행하여 파일의 특성들을 확인해볼 것입니다.


*: 실제 디스크에 저장되는 자료가 아닌, 소프트웨어 위의 논리 구조를 뜻합니다.

#### File types and permissions
UNIX 파일 시스템에서 파일의 종류와 권한은 10자리의 문자로 표시되며 (예: drwxr-xr-x), 각각은 다음과 같은 의미를 갖습니다.

|Filetype (1)|User (Permission) (3)|Group (3)|Others (3)|
|:---|:---:|:---:|:---:|
|d|rwx|r-x|r-x|                                         |

파일의 종류는 한 자리의 문자로 표시되며 8종류로 구별됩니다.

|Character (1)|Type      |
|:---|:---:|
|-|`Regular file`          |
|d|`Directory`             |
|D|Door file             |
|c|Character device file |
|b|Block device file     |
|s|Socket file           |
|p|FIFO (Named pipe) file|
|l|`Symbolic link file`    |

파일의 권한은 9자리 문자로 표시되며, 3자리씩 각각 사용자, 그룹, 외부에 해당하는 권한을 표시합니다. 각각의 권한은 동일한 표기법을 이용하며, 자릿수 순서대로 읽기 (r), 쓰기 (w), 실행 (x) 권한을 나타냅니다. 권한을 나타내는 표기법으로는 숫자 표기법과 문자 표기법이 있습니다. 다음 테이블은 첫번째 열에 숫자 표기법을, 나머지 열에 문자 표기법을 표시합니다.

|Numeric notation|Read (4)|Write (2)|Execute (1)|Description|
|:---:|:---:|:---:|:---:|:---:|
|0|-|-|-|No permissions                                     |
|1|-|-|x|**Only execute**                                   |
|2|-|w|-|**Only write**                                     |
|3|-|w|x|Write and execute                                  |
|4|r|-|-|**Only read**                                      |
|5|r|-|x|Read and execute                                   |
|6|r|w|-|Read and write                                     |
|7|r|w|x|Read, write, and execute                           |

파일의 권한은 파일을 소유하고 있는 소유자가 변경할 수 있습니다. 사용자, 그룹, 외부에 대해 모두 읽기만 가능하도록 변경한다면 다음 명령어를 이용합니다. 추후에 파일의 권한 변경을 통해 삭제가 어려운 파일을 만들어볼 것입니다.
```bash
chmod 444 filename
```

#### Filename
하나의 디렉토리 안의 디렉토리 항목들의 이름을 *파일명 (filename)* 이라고 부릅니다. 파일명에는 대부분의 문자를 사용할 수 있습니다. 사용할 수 없는 문자는 단 두 개로, *슬래시 (/)* 와 *NULL* 문자입니다. 슬래시는 하나의 *경로명 (pathname)* 을 구성하는 파일명들을 구분하는 역할을 하고, NULL 문자는 경로명의 끝을 표시합니다. NULL 문자는 컴퓨터 공학 관점에서 중요한 역할을 하지만, 본 부트캠프에서는 일단 *사용하기에 불편한 문자* 정도로만 기억해도 괜찮습니다. 그럼에도 불구하고, 파일명은 보통 $^*$사용자가 이해하기 편한 문자들만 사용하는 것이 권장됩니다. 파일명에 특수문자들이 포함되어 있으면 사용자의 언어 환경에 따라 읽고 쓰는 것이 상당히 복잡해질 수 있습니다.


$^{*}$: 운영체제 (혹은 시스템)간의 이식성을 위해 $^{**}$[POSIX](http://get.posixcertified.ieee.org/)는 파일명에 사용할 수 있는 문자를 영어(a-z, A-Z), 숫자(0-9), 마침표(.), 대쉬(-), 밑줄(_)로만 제한합니다.<br>
$^{**}$: POSIX: [IEEE (전기전자공학자협회)](https://www.ieee.org/)가 제정한 UNIX [응용 프로그래밍 인터페이스 (API)](https://en.wikipedia.org/wiki/API) 규격입니다. Linux 또한 UNIX-like 한 운영체제로써, 거의 모든 경우 POISX 규격을 따릅니다.

#### Pathname
경로명(pathname)은 하나 이상의 파일명들이 슬래시로 연결된 형태입니다. 경로명의 첫 글자가 / 일 수도 있으며, 이 경우 루트 디렉토리 이하의 모든 경로를 표시하는 것입니다. 이러한 표시 방식을 *절대 경로 (absolute path)* 혹은 *전체 경로 (full path)* 라고 합니다. 그렇지 않은 경우를 *상대 경로 (relative path)* 라고 하며, 이것은 현재 사용자의 디렉토리로부터 시작하는 경로의 상대적인 파일을 지칭합니다.

#### Special path
UNIX에는 새 디렉토리를 만들면 자동으로 생성되는 파일명이 두 개 있습니다. 하나는 **.**, 다른 하나는 **..** 입니다. .는 현재 디렉토리를 뜻하고, ..는 부모 디렉토리를 뜻합니다. 루트 디렉토리에서는 .와 ..가 동일하게 자기 자신인 루트 디렉토리를 뜻합니다. **~** 경로도 자주 사용됩니다. ~는 사용자의 홈 디렉토리입니다.

```bash
cd ..
```
```bash
cd snu-stat-bigdata-computing
```

#### Piping and redirection

| Concept      | Description                                                                                             | Example                                      |
|--------------|---------------------------------------------------------------------------------------------------------|----------------------------------------------|
| Pipe (&#124;)     | Connects two commands, sending the output of the first command as input to the second command.     | `command1` &#124; `command2`                        |
| Redirection (>)| Redirects the output of a command to a file, creating or overwriting the file.                    | `command > file.txt`                        |
| Redirection (>>) | Redirects the output of a command to a file, appending to the file's contents.                     | `command >> file.txt`                       |
| Redirection (<) | Uses the content of a file as input for a command.                                                | `command < input.txt`                      |
| Redirection (2>)| Redirects the error output (stderr) of a command to a file.                                      | `command 2> error.txt`                     |
| Redirection (&>)| Redirects both the standard output and error output of a command to a file.                      | `command &> output.txt`                    |
| Redirection (&>>) | Redirects both the standard output and error output of a command to a file, appending to the file. | `command &>> output.txt`                   |
| Redirection (2>&1)| Redirects the error output (stderr) of a command to the same location as standard output (stdout). | `command 2>&1`                              |

### Shell and profile

#### Login shell and non-login shell

Linux 환경에서 쉘은 로그인 쉘과 비 로그인 쉘로 나뉩니다. **로그인 쉘은 아이디와 패스워드를 입력하여 접속하는 쉘 환경**이고, **비 로그인 쉘은 로그인 없이 실행하는 쉘 환경입니다.** 예를 들어, 한번 로그인을 한 쉘 환경에서 또 쉘을 실행하는 경우 비 로그인 쉘이 실행됩니다.

로그인 쉘이 실행될 때, `~/.profile` 이란 파일을 참조하여 쉘이 실행됩니다.

```bash
cat ~/.profile
```
```
# ~/.profile: executed by the command interpreter for login shells.
# This file is not read by bash(1), if ~/.bash_profile or ~/.bash_login
# exists.
# see /usr/share/doc/bash/examples/startup-files for examples.
# the files are located in the bash-doc package.

# the default umask is set in /etc/profile; for setting the umask
# for ssh logins, install and configure the libpam-umask package.
#umask 022

# if running bash
if [ -n "$BASH_VERSION" ]; then
    # include .bashrc if it exists
    if [ -f "$HOME/.bashrc" ]; then
	. "$HOME/.bashrc"
    fi
fi

# set PATH so it includes user's private bin if it exists
if [ -d "$HOME/bin" ] ; then
    PATH="$HOME/bin:$PATH"
fi

# set PATH so it includes user's private bin if it exists
if [ -d "$HOME/.local/bin" ] ; then
    PATH="$HOME/.local/bin:$PATH"
fi
```

`~/.profile` 파일 내에는 `~/.bashrc` 라는 파일을 참조하는 부분이 있습니다. 따라서, `~/.profile` 이 실행될 때에는 반드시 `~/.bashrc` 가 참조됩니다.

비 로그인 쉘이 실행될 때에는 `~/.bashrc` 만을 참조하여 실행됩니다. `~/.profile` 혹은 `~/.bashrc` 내에는 개인의 환경 변수들과 정보를 작성할 수 있습니다. `~/.profile` 혹은 `~/.bashrc` 가 변경된 경우, 원칙적으로는 로그인 쉘 혹은 비 로그인 쉘을 다시 실행해야 변경 사항이 반영됩니다.

쉘을 재실행하지 않고 환경 설정을 적용할 수도 있습니다. 아래 명령어는 `~/.bashrc` 의 변경사항을 즉시 적용합니다.
```bash
source ~/.bashrc
```

긴 명령어의 경우 `alias` 를 이용하여 짧게 줄일 수 있습니다. 해당 설정을 유지하려면 `~/.profile`, `~/.bashrc` 와 같은 개인 프로필에 작성해두면 됩니다.
```bash
alias yy="ls -al"
```
```bash
echo "alias yy='ls -al'" >> ~/.profile
source ~/.profile
```

`~/.profile` 과 `~/.bashrc` 에는 개인의 환경 변수를 이용하여 쉘의 더 강력한 기능들을 활용할 수 있습니다. 보편적으로 환경 변수는 대문자로 작성합니다.
```bash
MYNAME="Yedarm Seong"
```
```bash
echo "MYNAME='Yedarm Seong'" >> ~/.bashrc && source ~/.bashrc
```

```bash
echo $MYNAME
```


각 열에 대한 내용은 다음과 같습니다. 특정 프로세스에 대해 접근할 때는 주로 PID (Process ID)를 이용하여 접근합니다.

| Column | Description                                             |
|--------|---------------------------------------------------------|
| UID    | User ID (The user who owns the process)                 |
| PID    | Process ID (The unique identifier of the process)       |
| PPID   | Parent Process ID (The ID of the parent process)        |
| C      | CPU Utilization (The percentage of CPU used by process) |
| STIME  | Start Time (The time when the process started)          |
| TTY    | Terminal Type (The terminal associated with the process)|
| TIME   | Total CPU Time (The total CPU time used by the process) |
| CMD    | Command (The command that started the process)          |

불필요한 프로세스, 혹은 과도한 자원을 프로세스 등은 프로세스 아이디를 이용하여 종료시킬 수 있습니다.
```bash
kill {PID}
```

다음 명령어를 통해 불필요한 프로그램을 실행한 후, `kill` 명령어를 이용해 종료합니다.
```bash
python do_not_run.py
```
터미널 환경을 하나 더 생성합니다.

```bash
kill 7461
```

`kill` 명령어는 사실 프로그램을 강제 종료하는 명령어가 아닙니다. `kill` 명령어는 주어진 프로세스에게 *신호 (signal)* 만을 전달합니다. 신호를 받은 프로세스가, 적절한 행동을 취하는 것입니다. 기본 신호는 15번 신호인 **SIGTERM**으로, 종료 요청을 전달합니다. 이 경우 프로그램은 가능한 안전한 형태로 종료하게 됩니다. 하지만, 프로그램의 상태에 따라 종료되지 않는 경우도 있습니다. 이러한 경우에는 가장 강력한 신호인 9번 신호는 **SIGKILL**을 사용합니다. SIGKILL은 프로세스를 강제 종료하도록 하는 신호입니다. 프로세스를 강제 종료할 경우 프로세스가 접근하던 데이터에 손상이 갈 수 있어 사용에 유의해야 합니다.


실행중인 프로세스에 대한 자세한 정보는 `/proc/[PID]` 디렉토리 내에서 확인할 수 있습니다. `/proc/[PID]` 내의 `cmdline` 에서는 프로세스를 실행한 명령어를 확인할 수 있고, `cwd`에서는 프로그램이 담긴 디렉토리에 대한 바로가기 링크 (=심볼릭 링크), `status`에는 `프로세스의 이름`, `상태`, `PID`, `PPID` 등 자세한 정보를 담고 있습니다.

터미널에 어떤 명령어를 입력하여 프로그램을 실행하였는지 확인할 수 있습니다.
```bash
cat /proc/{PID}/cmdline
```
실행중인 프로그램의 상태 정보들을 확인할 수 있습니다.
```bash
cat /proc/{PID}/status
```

### Useful commands

주어진 파일의 행의 개수, 단어의 수, 파일의 크기를 셀 수 있습니다. 단어의 수는 **공백 (space), 탭 (tab), 줄바꿈 문자 (newline)** 에 의해 분리됩니다.
```bash
wc {filename}
wc pattern1
```
```
11 105 591 pattern1
```

`wc` 명령어와 파이프를 응용하면 `ls` 등의 출력물의 결과에 대한 행의 수도 셀 수 있습니다. `-l` 옵션은 행의 수만 세는 옵션입니다.
```bash
ls -l | wc -l
```
```
16
```

```bash
more is less, less is more!
```
`more` 는 파일의 내용을 한번에 전부 가져와 출력해주는 명령어입니다. 한번에 많은 내용을 가져오기에 큰 파일의 경우 읽어오는 속도가 느립니다. 또한 파일의 이전 부분으로 이동할 수 없습니다.<br>
`less`는 파일의 일부분만을 먼저 보여주고, 방향키를 통해 파일의 이전, 다음 부분으로 이동할 수 있습니다. 이동할 때마다 해당 부분만을 읽어오기에 읽어오는 속도가 더 빠릅니다. 최근에는 두 명령어가 큰 차이가 없는 경우가 많습니다.

```bash
more {filename}
less {filename}
```
